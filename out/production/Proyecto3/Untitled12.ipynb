```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
  HAMBRE: CUANTA HAMBRE TIENE
  UMBRAL: A PARTIR DE CUANTA HAMBRE COME
  SACIEDAD:  CUANTA HAMBRE SE VA CADA VEZ QUE COME
"""
class Humano:

  def __init__(self, vida, umbral,saciedad, perdida_hambre):

    self.vida = vida
    self.hambre = 0
    self.umbral = umbral
    self.saciedad = saciedad
    self.perdida_hambre = perdida_hambre

    #perdida_hambre no debera ser parte saciedad o hambre? Es decir, si corre tiene una perdida_hambre de X, que sera hambre - X, o cmo sera?


  def comer (self):
    if self.hambre > self.umbral:
      self.hambre = self.hambre - self.saciedad
      if self.hambre > self.hambre + self.umbral: #No debera ser is self.hambre > self.umbral?
        self.vida += 5
      self.saciedad = (np.random.random_integers(5,30)) - max((- self.hambre) * 0.1,0)

  def vivir(self):
    self.hambre += self.perdida_hambre
    if self.hambre >= self.umbral:
      self.vida -= max(0 , (self.hambre - self.umbral ) / self.umbral)

  def estaVivo(self):
    if self.vida <=0:
      return False, "estas muerto"
    elif self.vida < 10:
      return True, "estas por morir"
    else:
      return True, ""


hambre = []
vida = []

epochs = 100

chs = Humano(120, 20, 5, 2)

for i in range(epochs):

  chs.vivir()

  hambre.append(chs.hambre)
  vida.append(chs.vida)


  chs.comer()


  estaVivo, msg = chs.estaVivo()
  if msg:
    print(msg)
  if not estaVivo:
    break


x = [i for i in range(len(hambre))]

fig, axs = plt.subplots(1, 2, figsize=(20,5))
axs[0].plot(x,hambre)
axs[0].scatter(x,hambre)
axs[0].hlines(20,0,100, color='r')
axs[1].plot(x,vida)
plt.show()
```
